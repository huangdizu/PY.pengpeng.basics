```python
#lesson25 Pandas資料分析-基礎教學
#載入pandas模組
import pandas as pd

#建立Series
data=pd.Series([20, 10 ,15])

#基本Series操作
print("Max", data.max())
print("Median", data.median())

#比較用法，一個一個跟20比較
data=data==20
print(data)
print("----------")

#建立Dataframe
data=pd.DataFrame({
    "name":["Amy", "John", "Bob"],
    "salary":[30000, 50000, 40000]
})
print(data)
print("----------")
print(data["salary"]) #取得特定的欄位
print("----------")
print(data.iloc[2]) #取得特定的列
```

```python
#lesson26 Pandas資料分析-Series單維度資料 
import pandas as pd

#資料索引
data=pd.Series([5, 4, -2, 3, 7], index=["a", "b", "c", "d", "e"])
print(data)
print("----------")

#觀察資料
print("資料型態", data.dtype)
print("資料數量", data.size)
print("資料索引", data.index)
print("----------")

#取得資料: 根據順序、根據索引
print(data[2], data[0])
print(data["e"], data["d"])
print("----------")

#數字運算: 基本、統計、順序
print("最大值", data.max())
print("總和", data.sum())
print("標準差", data.std())
print("中位數", data.median())
print("最大的三個數", data.nlargest(3), sep="\n")
print("最小的兩個數", data.nsmallest(2), sep="\n")
print("----------")

#字串運算: 基本、串接、搜尋、取代
data=pd.Series(["您好","Python","Pandas"])
print(data.str.upper()) #全部變大寫
print(data.str.len()) #算出每個字串的長度
print(data.str.cat(sep="--")) #把字串串起來，可以自訂串接的符號
print("----------")
print(data.str.contains("P")) #判斷每個字串是否包含特定的字元
print(data.str.replace("您好", "Hello"))
#注意下面這個就算上面字串replace還是舊的data呦
print(data)
```

```python
#lesson27 Pandas資料分析-DataFrame雙維度資料
import pandas as pd

#資料索引 data=pd.DataFrame(字典，index=索引列表)
data=pd.DataFrame({
    "name":["Amy", "Bob", "Charles"],
    "salary":[30000, 50000, 40000]
}, index=["a", "b", "c"])
print(data)
print("----------")

#觀察資料
print("資料型態(列,欄)", data.shape)
print("資料數量", data.size)
print("資料索引", data.index)
print("----------")

#取得Row的Series資料: 根據順序、根據序列
print("取得第二列", data.iloc[1], sep="\n")
print("取得第c列", data.loc["c"], sep="\n")
print("----------")

#取得Column的Series資料: 根據欄位的名稱
print("取得name欄位", data["name"], sep="\n")
print("----------")

names=data["name"] #取得單維度的Series資料
print("把name全部轉大寫", names.str.upper(), sep="\n")
print("----------")

salaries=data["salary"]
print("薪水的平均值", salaries.mean())
print("----------")

#建立新的欄位
data["revenue"]=[500000, 400000, 300000] #data[新欄位的名稱]=列表
data["rank"]=pd.Series([3,6,1], index=["a", "b", "c"]) #data=[新欄位的名稱]=Series的資料
data["cp"]=data["revenue"]/data["salary"]
data["dog"]=data["revenue"]
print(data)
```

```python
#lesson28 Pandas資料分析-篩選資料
import pandas as pd
#篩選練習-Series
data=pd.Series([30, 15, 20])
condition=data>18
print(condition) #印出來可以看condition是怎麼運作的
filteredData=data[condition]
print(filteredData)
print("----------")

data=pd.Series(["您好","Python","Pandas"])
condition=data.str.contains("P") #試試看("[P好]")和("P好")，注意其中的小細節
filteredData=data[condition]
print(filteredData)
print("----------")

#篩選練習-DataFrame
data=pd.DataFrame({
    "name":["Amy", "Bob", "Charles"],
    "salary":[30000, 50000, 40000]
})
print(data)
condition=data["salary"]>=40000
print(condition) #印出來可以看condition是怎麼運作的
filteredData=data[condition]
print(filteredData)
print("----------")

condition=data["name"]=="Amy"
filteredData=data[condition]
print(filteredData)
```

```python
#lesson29 Pandas資料分析-GooglePlayStore資料集
import pandas as pd

#讀取資料
data=pd.read_csv("C:/Users/eason/Desktop/python彭彭/googleplaystore.csv") #把csv格式的檔案讀取成DataFrame

#觀察資料
print("資料型態", data.shape)
print("資料欄位", data.columns)
print("----------")

#分析資料:評分的各種統計數據
print("平均數", data["Rating"].mean())
print("中位數", data["Rating"].median())
print("取得前一百個應用程式的平均", data["Rating"].nlargest(100).mean())
print("----------")

#跑出來後發現:照理說google的評分不應該超過五分，因此我們要找出原因
condition=data["Rating"]>5
datatry=data[condition]
print(datatry) #得知10472號評分不合理
print("----------")

#因此應該這樣解
condition=data["Rating"]<=5
data=data[condition]
print("平均數", data["Rating"].mean())
print("中位數", data["Rating"].median())
print("取得前一百個應用程式的平均", data["Rating"].nlargest(100).mean())
print("----------")

#分析資料:安裝數量的各種統計數據
print(data["Installs"]) #我們觀察到原始的資料裡面，不是數字，而是字串的型態(dtype:object)
#因此我們用to_numeric試著將字串轉換成數字
#因為有,+在字串裡面，所以無法將其轉成數字，因此我們用str.replace
#後來又發現在某一筆資料裡有"Free"這個奇怪字串，因此我們用replace將其取代
print("----------")
data["Installs"]=pd.to_numeric(data["Installs"].str.replace("[,+]","").replace("Free","")) 
print("平均數", data["Installs"].mean())
print("----------")
condition=data["Installs"]>100000
print(data[condition])
print("----------")
print("安裝數量大於100000的應用程式有幾個", data[condition].shape[0]) #shape可以算出數量，因為其顯示行和列，而[0]可以選擇保留(4947, 13)中的4947
print("----------")

#基於資料的應用:關鍵字搜尋應用程式名稱 
keyword=input("請輸入關鍵字: ")
condition=data["App"].str.contains(keyword, case=False) #case=False可忽略大小寫
print("包含關鍵字的應用程式數量", data[condition].shape[0])
print(data[condition]["App"])
```
