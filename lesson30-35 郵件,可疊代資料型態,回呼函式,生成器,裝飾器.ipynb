```python
#lesson31 可疊代資料型態-Iterable Data Types
#Iterable資料型態: 字串、列表、集合、字典

#for迴圈
#for 變數名稱 in 可疊代的資料:
#以集合為例
set={"a", "test", 3, 10} #set沒有順序性
for x in set:
    print(x)
print("----------")

#以字典為例
dic={"a":3, "x":5}
for key in dic:
    print(key) #只印出key
print("----------")
    
dic={"a":3, "x":5}
for key in dic:    
    print(dic[key]) #可以印出對應的value
print("----------")

#內建函式
#max(可疊代資料)
result=max([10, 2, 30, 1])
print(result)
result=max("xaz")
print(result)
result=max({10, 200, 30, 1}) #set依舊可以辨別大小
print(result)
result=max({"a":3, "x":5}) #注意: 是用key在辨別大小
print(result)
print("----------")

#sorted(可疊代資料)
result=sorted("cab")
print(result)
result=sorted({10, 2, 200, -5})
print(result)
result=sorted({"a":3, "x":5}) #注意: 是用key在辨別大小
print(result)
```

```python
#lesson32 生成器Generator基礎使用

#我有一個疑問: 生成器是一個函式丟給一個變數??? 不知道是不是print和yield的差別，print好像是直接印出來，yield就只是生成在那邊

#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#這個是我自己寫的，比較dog和generateEven
def generateEven(maxNumber):
    number=0
    while number<=maxNumber:
        yield(number)
        number+=2
print(generateEven)
dog=generateEven(100)
print(dog)
print("----------")
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

#定義生成器函式
def test():
    print("階段一")
    yield 5
    print("階段二")
    yield 10
#呼叫並回傳生成器
gen=test()
print(gen) #可以得知，你可以馬上生成一個生成器，而不是產出一個5
#搭配for迴圈使用
for d in gen:
    print(d)
print("----------")

#有最大數量限制的
def generateEven(maxNumber):
    number=0
    while number<=maxNumber:
        yield number
        number+=2
evenGenerator=generateEven(30)
for data in evenGenerator:
    print(data)
print("----------")

#這個會無限輪迴，想看要把最後的兩個#刪掉
def generateEven():
    number=0
    while True:
        yield number
        number+=2
evenGenerator=generateEven()
#for data in evenGenerator:
#    print(data)
```

```python
#lesson34 裝飾器Decorator-基本篇

#為什麼一定要return run????

#定義裝飾器
def myDeco(callback):
    def run():
        print("裝飾器中的程式碼")
        callback(987) #這個回呼函式，其實就是被裝飾的普通函式
    return run
#使用裝飾器
@myDeco 
def test(arg):
    print("普通函式的程式碼", arg)
test()
print("----------")


#使用裝飾器的意義: 會先執行裝飾器想要做的事情，再來執行本來的函式
#用裝飾器就可以選擇，最後print出的結果，要是英文或中文都可以
#定義一個裝飾器:計算1+2+...+50的總和
def calculate(callback):
    def run():
        #裝飾器想要執行的程式碼
        result=0
        for n in range(51):
            result+=n
        callback(result)
    return run
#使用裝飾器
@calculate
def show(n):
    print("計算結果是", n)
@calculate
def showEnglish(n):
    print("Result is", n)
show()
showEnglish()
```

```python
#lesson35 裝飾器工廠Decorator Factory
#裝飾器工廠可以接受額外的參數，讓裝飾器內部的程式使用這個參數，參數即成為有彈性的部分
def myFactory(base):
    def myDeco(callback):
        def run():
            print("裝飾器內的程式", base)
            result=base*2
            callback(result)
        return run
    return myDeco

@myFactory([1,3,6,8]) #括號內可以放入想放的東西
def test(dog):
    print("普通函式的程式", dog)

test()
print("----------")


#計算1+2+...+max的裝飾器
#用裝飾器就可以不要只加到10，可以更換參數想加到多少就多少
def calculateFactory(max):
    def calculate(callback):
        def run():
            sum=0
            for n in range(max):
                sum+=n
            callback(sum)
        return run
    return calculate

@calculateFactory(6543)
def show(result):
    print("加起來的總合: ", result)
@calculateFactory(87654)
def showEnglish(result):
    print("The result is ", result)

show() #補充 括號內的參數對應到run()的參數，如果有放的話啦
showEnglish()
```
